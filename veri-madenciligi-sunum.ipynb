# ğŸ§‘â€ğŸ’» KNN AlgoritmasÄ±

KNN (K-Nearest Neighbors) algoritmasÄ±, bir sÄ±nÄ±flandÄ±rma yÃ¶ntemidir ve yeni verilerin hangi sÄ±nÄ±fa ait olduÄŸunu, eÄŸitim verisindeki en yakÄ±n komÅŸularÄ±na gÃ¶re tahmin eder. Bu algoritma, mesafe Ã¶lÃ§Ã¼leri kullanarak, yeni veri noktasÄ±nÄ± eÄŸitim verisindeki benzer noktalarla karÅŸÄ±laÅŸtÄ±rÄ±r ve en yakÄ±n k komÅŸusunun etiketine gÃ¶re tahmin yapar.

```python
# KNN algoritmasÄ±
class KNN:
    def __init__(self, k=3):
        self.k = k
        self.X_train = []
        self.y_train = []

    def fit(self, X, y):
        # EÄŸitim verilerini sakla
        self.X_train = X
        self.y_train = y

    def predict(self, X):
        # Yeni veri setindeki her bir Ã¶rnek iÃ§in tahmin yap
        predictions = []
        for x in X:
            predictions.append(self._predict_single_point(x))
        return predictions

    def _predict_single_point(self, x):
        # Mesafeleri hesapla
        distances = []
        for i, train_point in enumerate(self.X_train):
            distance = self._euclidean_distance(train_point, x)
            distances.append((distance, self.y_train[i]))

        # En kÃ¼Ã§Ã¼k mesafeye gÃ¶re sÄ±ralama
        distances.sort(key=lambda d: d[0])

        # En yakÄ±n k komÅŸunun etiketlerini al
        k_nearest_labels = [distances[i][1] for i in range(self.k)]

        # En sÄ±k gÃ¶rÃ¼len etiketi bul
        return self._most_common_label(k_nearest_labels)

    def _euclidean_distance(self, point1, point2):
        # Ä°ki nokta arasÄ±ndaki Ã–klid mesafesi
        distance = 0
        for a, b in zip(point1, point2):
            distance += (a - b) ** 2
        return distance ** 0.5

    def _most_common_label(self, labels):
        # En sÄ±k gÃ¶rÃ¼len etiketi bul
        label_count = {}
        for label in labels:
            if label not in label_count:
                label_count[label] = 1
            else:
                label_count[label] += 1
        # En yÃ¼ksek sayÄ±ya sahip etiketi dÃ¶ndÃ¼r
        return max(label_count, key=label_count.get)
```

## ğŸ” En Ä°yi N DeÄŸeri (k) HesaplanÄ±r

Bu aÅŸamada, KNN algoritmasÄ± iÃ§in en iyi sonuÃ§ veren k deÄŸeri (komÅŸu sayÄ±sÄ±) hesaplanÄ±r. FarklÄ± k deÄŸerleri denenerek her birinin doÄŸruluk oranÄ± deÄŸerlendirilir ve en yÃ¼ksek doÄŸruluÄŸa sahip k deÄŸeri seÃ§ilir.

```python
best_k = 1
best_accuracy = 0
k_values = range(1, 21)  # k=1'den k=5'e kadar deÄŸerlendir
accuracies = []

for k in k_values:
    knn = KNN(k=k)
    knn.fit(X_train_scaled, y_train)  # y_train zaten NumPy formatÄ±nda
    y_pred = knn.predict(X_test_scaled)  # X zaten NumPy formatÄ±nda
    accuracy = accuracy_score(y_test, y_pred)
    accuracies.append(accuracy_score(y_test, y_pred))

    print(f"k = {k}, Accuracy = {accuracy:.2f}")
    if accuracy > best_accuracy:
        best_k = k
        best_accuracy = accuracy

print(f"\nBest k: {best_k}, Best Accuracy: {best_accuracy:.2f}")

# En iyi k ile modeli yeniden eÄŸitme
knn = KNN(k=best_k)
knn.fit(X_train_scaled, y_train)  # y_train zaten NumPy formatÄ±nda
y_pred = knn.predict(X_test_scaled)
```

# ğŸ“ˆ Performans Ã‡Ä±ktÄ±sÄ±

Modelin performansÄ±nÄ± deÄŸerlendirmek iÃ§in Ã§eÅŸitli metrikler kullanÄ±lÄ±r. Bu aÅŸamada, en iyi k deÄŸeriyle yapÄ±lan tahminlerin ardÄ±ndan sÄ±nÄ±flandÄ±rma raporu gÃ¶rÃ¼ntÃ¼lenir.

```python
# Performans raporu
print("\nClassification Report (Best k):")
print(classification_report(y_test, y_pred))
```

Modelin doÄŸruluÄŸunu gÃ¶rselleÅŸtirmek iÃ§in bir Ã§izelge oluÅŸturulur. Bu Ã§izelge, k deÄŸerinin (komÅŸu sayÄ±sÄ±) doÄŸruluk Ã¼zerindeki etkisini gÃ¶sterir. AyrÄ±ca, doÄŸruluk deÄŸerlerinin etrafÄ±nda bir standart sapma bÃ¶lgesi eklenir.

# ğŸ“‰ PCA ile Boyut Ä°ndirgeme

PCA (Principal Component Analysis) kullanÄ±larak, yÃ¼ksek boyutlu verinin daha dÃ¼ÅŸÃ¼k bir boyutta gÃ¶rselleÅŸtirilmesi saÄŸlanÄ±r. Bu iÅŸlem, verinin daha iyi anlaÅŸÄ±lmasÄ±nÄ± ve analiz edilmesini kolaylaÅŸtÄ±rÄ±r.

Ä°lk olarak, `PCA` fonksiyonu ile verinin iki bileÅŸene indirgenmesi yapÄ±lÄ±r. ArdÄ±ndan, bu iki bileÅŸenin gÃ¶rselleÅŸtirilmesi iÃ§in bir scatter plot (daÄŸÄ±lÄ±m grafiÄŸi) oluÅŸturulur.

```python
# PCA ile boyut indirgeme
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X_train_scaled)
X_test_reduced = pca.transform(X_test_scaled)
# PCA GÃ¶rselleÅŸtirme
plt.figure(figsize=(10, 6))
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y_train, cmap=ListedColormap(['#FF0000', '#0000FF']), edgecolor='k')
plt.title("PCA ile Veri GÃ¶rselleÅŸtirme")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal ComponentÂ 2")
plt.show()
```

# ğŸ¨ PCA ile Boyut Ä°ndirgeme ve KNN Karar SÄ±nÄ±rlarÄ±

Bu gÃ¶rselleÅŸtirme, PCA ile boyutlarÄ± indirgenmiÅŸ veride KNN algoritmasÄ±nÄ±n karar sÄ±nÄ±rlarÄ±nÄ± gÃ¶sterir. 

1. **KNN Modeli EÄŸitimi**: 
   - `X_reduced` (indirgenmiÅŸ Ã¶zellikler) ve `y_train` (etiketler) kullanÄ±larak model eÄŸitilir. 

2. **Karar SÄ±nÄ±rlarÄ± Ä°Ã§in Mesh Grid OluÅŸturma**:
   - Veri aralÄ±ÄŸÄ±nÄ± kapsayan bir mesh grid (kafes Ä±zgarasÄ±) oluÅŸturulur. Bu grid, KNN'in karar sÄ±nÄ±rlarÄ±nÄ± tahmin edebilmesi iÃ§in kullanÄ±lÄ±r. ğŸ”

3. **Karar SÄ±nÄ±rlarÄ±nÄ±n HesaplanmasÄ±**:
   - Mesh grid Ã¼zerindeki her bir noktada modelin tahminleri alÄ±nÄ±r ve karar sÄ±nÄ±rlarÄ±nÄ± belirlemek iÃ§in bu tahminler gÃ¶rselleÅŸtirilir.

4. **Veri ve Karar SÄ±nÄ±rlarÄ±nÄ±n GÃ¶rselleÅŸtirilmesi**:
   - Karar sÄ±nÄ±rlarÄ±, iki farklÄ± renk ile gÃ¶sterilir. 
   - EÄŸitim verisi ise renkli noktalar olarak iÅŸaretlenir:
     - ğŸ”´ KÄ±rmÄ±zÄ±: Riskli birey
     - ğŸ”µ Mavi: Risksiz birey

5. **Grafik DetaylarÄ±**:
   - X ve Y eksenlerinde, PCA ile indirgenmiÅŸ ana bileÅŸenler yer alÄ±r.
   - KNN modelinin parametrelerinden biri olan **k** deÄŸeri (komÅŸu sayÄ±sÄ±) grafikte belirtilmiÅŸtir.

Bu gÃ¶rselleÅŸtirme, sÄ±nÄ±flandÄ±rma algoritmasÄ±nÄ±n kararlarÄ±nÄ± daha iyi anlamamÄ±zÄ± ve veri ile algoritmanÄ±n uyumunu gÃ¶rmemizi saÄŸlar. ğŸ“Š


```python
knn.fit(X_reduced, y_train)

# Karar sÄ±nÄ±rlarÄ± iÃ§in mesh grid oluÅŸtur
x_min, x_max = X_reduced[:, 0].min() - 1, X_reduced[:, 0].max() + 1
y_min, y_max = X_reduced[:, 1].min() - 1, X_reduced[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))

# Mesh grid Ã¼zerindeki tahminleri yap
Z = np.array(knn.predict(np.c_[xx.ravel(), yy.ravel()]))
Z = Z.reshape(xx.shape)

# Karar sÄ±nÄ±rlarÄ±nÄ± ve veriyi gÃ¶rselleÅŸtir
plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, alpha=0.6, cmap=ListedColormap(['#FFAAAA', '#AAAAFF']))  # Karar sÄ±nÄ±rlarÄ±
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y_train, cmap=ListedColormap(['#FF0000', '#0000FF']), edgecolor='k')  # EÄŸitim verisi
plt.title(f"PCA ile Boyut Ä°ndirgeme ve KNN Karar SÄ±nÄ±rlarÄ± (k={best_k})")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```
